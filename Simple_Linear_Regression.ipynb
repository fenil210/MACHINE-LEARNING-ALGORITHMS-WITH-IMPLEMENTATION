Single Variable Linear Regression

Predict Price of Area per Square feet

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

#url="https://github.com/DarjiKrisha/Python/blob/main/print.csv"
#names=['area','price']
#dataset=pd.read_csv(url,names=names)

ls1=[2000,3000,3200,3600,4000]
ls2=[550000,565000,610000,680000,725000]
plt.scatter(ls1,ls2)
#plt.plot(ls1,reg.predict)
```

```python
#wine=np.asarray(ls1)
reg = linear_model.LinearRegression()
X=[]
for x in ls1:
  ls=[x]
  X.append(ls)
print(X)
reg.fit(X,ls2)
plt.scatter(ls1,ls2,color="red")
plt.plot(ls1,reg.predict(X),color='blue')
```

Multivariate Regression


Home Prices 

```python
lst1=[[2000,3,20],[3000,4,15],[3200,None,18],[3600,3,30],[4000,5,8]]
lst2=[550000,565000,610000,595000,760000]

import math
med=0.0
for x in lst1:
  for i in range(len(x)):
    if(i==1 and x[i]!=None):
      med=med+x[i]
med=math.floor(med/(len(lst1)))
#print(math.floor(med))
for x in lst1:
  for i in range(len(x)):
    if(x[i]==None):
      x[i]=med
print(lst1)
```

```python
reg1= linear_model.LinearRegression()
reg1.fit(lst1,lst2)
```

```python
print(reg1.coef_)
print(reg1.intercept_)
```

```python
reg1.predict([[2500,3,15]])
reg1.predict([[3500,4,9]])
```
